#### 1.4 검색기(retriever)활용

```python
from langchain.chat_models import ChatOpenAI
```

```python
llm = ChatOpenAI(model_name="gpt-4o-mini",
                 temperature=0
                 )
```

##### chain_type: 검색된 문서를 언어 모델에 전달하여 답변을 생성하는 방식을 결정하는 파라미터입니다. 주로 다음과 같은 값들이 사용됩니다.
 - "stuff": 검색된 모든 문서를 하나의 프롬프트에 담아 언어 모델에 전달합니다. 문서의 수가 적고 전체 내용을 한 번에 처리할 수 있을 때 효율적입니다.
 - "map_reduce": 검색된 문서를 여러 개의 프롬프트로 나누어 언어 모델에 전달하고, 각 프롬프트에 대한 답변을 요약하여 최종 답변을 생성합니다. 문서의 수가 많거나 각 문서의 내용이 길 때 유용합니다.
 - "refine": 검색된 문서를 순차적으로 언어 모델에 전달하면서 답변을 점진적으로 개선해 나갑니다. 초기 답변을 기반으로 추가 정보를 통합하여 최종 답변의 정확도를 높일 수 있습니다.
 - "map_rerank": 각 문서에 대한 답변의 관련성을 평가하고, 가장 관련성이 높은 답변을 선택합니다.

```python
from langchain.chains import RetrievalQA

retriever = db_faiss.as_retriever()


qa_retrieval = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever)

'''
chain_type: 검색된 문서를 언어 모델에 전달하여 답변을 생성하는 방식을 결정하는 파라미터입니다. 주로 다음과 같은 값들이 사용됩니다.
"stuff": 검색된 모든 문서를 하나의 프롬프트에 담아 언어 모델에 전달합니다. 문서의 수가 적고 전체 내용을 한 번에 처리할 수 있을 때 효율적입니다.
"map_reduce": 검색된 문서를 여러 개의 프롬프트로 나누어 언어 모델에 전달하고, 각 프롬프트에 대한 답변을 요약하여 최종 답변을 생성합니다. 문서의 수가 많거나 각 문서의 내용이 길 때 유용합니다.
"refine": 검색된 문서를 순차적으로 언어 모델에 전달하면서 답변을 점진적으로 개선해 나갑니다. 초기 답변을 기반으로 추가 정보를 통합하여 최종 답변의 정확도를 높일 수 있습니다.
"map_rerank": 각 문서에 대한 답변의 관련성을 평가하고, 가장 관련성이 높은 답변을 선택합니다.
'''

query = "마을 무덤에 있던 남자를 죽인 사람은 누구냐"
result = qa_retrieval({"query": query})

print(result['result'])
```

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate
from langchain.chat_models import ChatOpenAI

from dotenv import load_dotenv

load_dotenv()

```

```python
llm = ChatOpenAI(model_name="gpt-4o-mini",
                 temperature=0,
                )

prompt = PromptTemplate(
    input_variables=["country"],
    template = 'What city is the capital of {country}?'
)

chain = LLMChain(llm=llm, prompt=prompt) #프롬프트와 모델을 체인으로 연결
chain.run("North Korea")
```

```python
# define prompt1

prompt1 = PromptTemplate(
    input_variables=["sentence"],
    template = 'Please trasnlate the sentence into Korean: {sentence}?'
)

# translate chain1 model

chain1 = LLMChain(llm=llm, prompt=prompt1, output_key="translation")
```

```python
# define prompt2

prompt2 = PromptTemplate.from_template(
    template = '다음문장을 한문장으로 요약하세요: {translation}?'
)

# summary chain2 model
chain2 = LLMChain(llm = llm, prompt = prompt2, output_key = 'summary')

from langchain.chains import SequentialChain

all_chain = SequentialChain(
    chains = [chain1, chain2],
    input_variables = ['sentence'],
    output_variables = ['translation', 'summary']
)
```

```python
sentence = '''Tom Sawyer lived with his aunt because his mother and 
father were dead. Tom didn’t like going to school, and he 
didn’t like working. He liked playing and having 
adventures. One Friday, he didn’t go to school—he went 
to the river. 
Aunt Polly was angry. “You’re a bad boy!” she said. 
“Tomorrow you can’t play with your friends because you 
didn’t go to school today. Tomorrow you’re going to work 
for me. You can paint the fence.” 
Saturday morning, Tom was not happy, but he started to 
paint the fence. His friend Jim was in the street. 
Tom asked him, “Do you want to paint?” 
Jim said, “No, I can’t. I’m going to get water.” 
Then Ben came to Tom’s house. He watched Tom and 
said, “I’m going to swim today. You can’t swim because 
you’re working.” '''
```

```python
print(all_chain(sentence))
```

```python
from langchain.chat_models import ChatOpenAI

llm = ChatOpenAI(model_name="gpt-4o-mini",
                 temperature=0,
                )


from langchain.chains import ConversationChain
conversation = ConversationChain(
    llm=llm,
    verbose=True
)
# verbose = True 대화체인이 작동하는 동안 내부적으로 어떤 일이 일어나는지 출려 됨, 중간단계를 포함
# verbose = False 대화 체인이 작동하는 동안 중간 단계를 생략함


conversation.predict(input="진희는 강아지를 한마리 키우고 있습니다")
conversation.predict(input="영수는 고양이를 두마리 키우고 있습니다")
conversation.predict(input="진희와 영수가 키우고 있는 동물은 몇마리 일까요?")

```
